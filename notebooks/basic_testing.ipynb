```python
%cd ..
from pygoa import GOSnapshot, to_df
```

```python
from datetime import date
import sys

summaries = []


for year in [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]:
    for d in [date(year, m, 1) for m in range(1, 13)]:
        try:
            sp = GOSnapshot(date=d.strftime("%Y-%m-%d"))
            print(sp.summary.nterms, sp.summary.nchildren)
            summaries.append(sp.summary)
        except IOError:
            print("File not available for date:", d, sys.exc_info())
        except:
            print(sys.exc_info())
            
```

```python
df = to_df(summaries)
```
